```python
# ==============================================================
# 🔥 TRANSFORMER MODEL FOR BROILER PRICE PREDICTION
# ==============================================================

# ==============================================================
# 📂 STEP 1 — Upload CSV
# ==============================================================

from google.colab import files
import io
import pandas as pd

print("📂 Upload your CSV file:")
uploaded = files.upload()

filename = list(uploaded.keys())[0]

df_raw = pd.read_csv(io.BytesIO(uploaded[filename]))

print(f"✅ Loaded: {filename}")
display(df_raw.head())


# ==============================================================
# 📦 STEP 2 — Imports
# ==============================================================

# !pip install tensorflow -q

import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

import tensorflow as tf

from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input,
    Dense,
    Dropout,
    LayerNormalization,
    MultiHeadAttention,
    GlobalAveragePooling1D
)

print("✅ Libraries loaded")


# ==============================================================
# 🧹 STEP 3 — Clean Data
# ==============================================================

df = df_raw.copy()

df["date"] = pd.to_datetime(df["date"], errors="coerce")

df = (
    df.dropna(subset=["date"])
      .sort_values("date")
      .reset_index(drop=True)
)

# Use important features
df = df[[
    "date",
    "avg_farm_rate",
    "avg_doc",
    "egg_per_30doz"
]].dropna()

print("✅ Cleaned data:", df.shape)


# ==============================================================
# 🔧 STEP 4 — Scaling
# ==============================================================

features = [
    "avg_farm_rate",
    "avg_doc",
    "egg_per_30doz"
]

scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(df[features])

print("✅ Data scaled")


# ==============================================================
# 🔄 STEP 5 — Create Sequences
# ==============================================================

SEQ_LEN = 30

X = []
y = []

for i in range(SEQ_LEN, len(scaled_data)):

    X.append(
        scaled_data[i-SEQ_LEN:i]
    )

    # predict price only
    y.append(
        scaled_data[i][0]
    )

X = np.array(X)
y = np.array(y)

print("✅ Sequence shape:", X.shape)


# ==============================================================
# ✂️ STEP 6 — Train/Test Split
# ==============================================================

split = int(len(X) * 0.8)

X_train = X[:split]
y_train = y[:split]

X_test = X[split:]
y_test = y[split:]

print("✅ Train/Test split ready")


# ==============================================================
# 🧠 STEP 7 — BUILD TRANSFORMER
# ==============================================================

def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):

    # Attention
    x = MultiHeadAttention(
        key_dim=head_size,
        num_heads=num_heads,
        dropout=dropout
    )(inputs, inputs)

    x = Dropout(dropout)(x)

    x = LayerNormalization(epsilon=1e-6)(
        x + inputs
    )

    # Feed Forward
    ff = Dense(ff_dim, activation="relu")(x)

    ff = Dropout(dropout)(ff)

    ff = Dense(inputs.shape[-1])(ff)

    return LayerNormalization(epsilon=1e-6)(
        x + ff
    )


# Input
inputs = Input(shape=(SEQ_LEN, X.shape[2]))

# Transformer blocks
x = transformer_encoder(
    inputs,
    head_size=64,
    num_heads=4,
    ff_dim=128,
    dropout=0.2
)

x = transformer_encoder(
    x,
    head_size=64,
    num_heads=4,
    ff_dim=128,
    dropout=0.2
)

# Pooling
x = GlobalAveragePooling1D()(x)

# Dense layers
x = Dense(64, activation="relu")(x)

x = Dropout(0.2)(x)

outputs = Dense(1)(x)

# Final model
model = Model(inputs, outputs)

model.compile(
    optimizer="adam",
    loss="mse"
)

model.summary()


# ==============================================================
# 🏋️ STEP 8 — Train Transformer
# ==============================================================

history = model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=32,
    validation_data=(X_test, y_test)
)


# ==============================================================
# 📊 STEP 9 — Evaluate
# ==============================================================

preds = model.predict(X_test)

# inverse scaling
preds_full = np.zeros((len(preds), 3))
preds_full[:, 0] = preds[:, 0]

y_full = np.zeros((len(y_test), 3))
y_full[:, 0] = y_test

preds_actual = scaler.inverse_transform(preds_full)[:, 0]
y_actual = scaler.inverse_transform(y_full)[:, 0]

mae = mean_absolute_error(
    y_actual,
    preds_actual
)

rmse = np.sqrt(
    mean_squared_error(
        y_actual,
        preds_actual
    )
)

print("\n📊 TRANSFORMER RESULTS")
print("=" * 40)

print(f"MAE  : {mae:.2f}")
print(f"RMSE : {rmse:.2f}")


# ==============================================================
# 🔮 STEP 10 — Future Prediction
# ==============================================================

def predict_future_transformer(
    model,
    scaled_data,
    days=7
):

    temp = scaled_data[-SEQ_LEN:].copy()

    preds = []

    for _ in range(days):

        x = temp.reshape(
            1,
            SEQ_LEN,
            temp.shape[1]
        )

        pred = model.predict(
            x,
            verbose=0
        )[0][0]

        # drift control
        last_val = temp[-1][0]

        pred = max(
            pred,
            last_val - 0.03
        )

        pred = min(
            pred,
            last_val + 0.03
        )

        preds.append(pred)

        # new row
        new_row = temp[-1].copy()

        new_row[0] = pred

        temp = np.vstack([
            temp[1:],
            new_row
        ])

    # inverse scaling
    preds_full = np.zeros(
        (len(preds), 3)
    )

    preds_full[:, 0] = preds

    return scaler.inverse_transform(
        preds_full
    )[:, 0]


# ==============================================================
# 🔮 STEP 11 — Predict Future
# ==============================================================

days = int(
    input(
        "\nEnter number of future days to predict: "
    )
)

if days > 14:
    print(
        "⚠️ Long-term predictions may be less accurate"
    )

future_preds = predict_future_transformer(
    model,
    scaled_data,
    days
)

print("\n📊 Future Predictions")
print("=" * 40)

for i, val in enumerate(future_preds):

    print(
        f"Day +{i+1}: PKR {val:.1f}"
    )


# ==============================================================
# 📈 STEP 12 — Plot Forecast
# ==============================================================

future_dates = pd.date_range(
    df["date"].iloc[-1],
    periods=days+1
)[1:]

plt.figure(figsize=(13,5))

plt.plot(
    df["date"],
    df["avg_farm_rate"],
    label="Historical"
)

plt.plot(
    future_dates,
    future_preds,
    '--',
    linewidth=2,
    label="Transformer Forecast"
)

plt.title(
    "Transformer Future Forecast"
)

plt.xlabel("Date")

plt.ylabel("PKR/kg")

plt.legend()

plt.show()
```
